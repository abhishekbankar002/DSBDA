```python
print("Old shape ", df.shape)

Q1 = np.percentile(df['Second year:   Sem 1'], 25 )
 
Q3 = np.percentile(df['Second year:   Sem 1'], 75 )
IQR = Q3 - Q1
# Upper bound
upper = np.where(df['Second year:   Sem 1'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['Second year:   Sem 1'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)

# print(lower[0], Q3, Q1)

print("New shape ", df.shape)
```

```python
from sklearn import preprocessing

min_max_scaler= preprocessing.MinMaxScaler()
df_normalized = min_max_scaler.fit_transform(df.iloc[:,7:11])
df_normalized = pd.DataFrame(df_normalized, columns=['FE Sem1','FE Sem2','SE Sem1','SE Sem2'])

df_normalized
```
