```python
lower_bound=.69
upper_bound=.991
res=df['Second year:   Sem 1'].quantile([lower_bound,upper_bound])
res
```

```python
trueindex=(res.loc[lower_bound]<df['Second year:   Sem 1'].values) & (df['Second year:   Sem 1'].values<res.loc[upper_bound])
trueindex
```

```python
falseindex=~trueindex
falseindex
```

```python
df['Second year:   Sem 1'][trueindex]
```
