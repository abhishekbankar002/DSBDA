```python
from sklearn import preprocessing
```

```python
df = iris
```

```python
df.head()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
df_normalized = min_max_scaler.fit_transform(df.iloc[:,[0,1,2,3]])
```

```python
df_normalized = pd.DataFrame(df_normalized, columns = ['sepal_length','sepal_width','petal_length','petal_width'])
df_normalized
```

### Categorical Values

```python
df['species'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])
```

```python
df['species'].unique()
df
```

### Sklearn for One hot encoding

```python
df['species'].unique()
```

```python
features_df = df.drop(columns=['species'])
```

```python
enc = preprocessing.OneHotEncoder()
```

```python
y = enc.fit_transform(df[['species']]).toarray()
```

```python
enc_df = pd.DataFrame(y)
```

```python
enc_df.head()
```

```python
df_encode = features_df.join(enc_df)
```

```python
df_encode.head()
```
